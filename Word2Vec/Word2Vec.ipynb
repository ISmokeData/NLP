```python
import gensim
import os
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('Data'):
  print(filename)
  # The with open statement should be indented within the for loop
  with open(os.path.join('Data', filename), encoding='utf-8', errors='ignore') as f:
      corpus = f.read()
      raw_sent = sent_tokenize(corpus)
      for sent in raw_sent:
          story.append(simple_preprocess(sent))
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('data'):
    f = open(os.path.join('data', filename),encoding='utf-8', errors='ignore')
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2, # accept those sentence which have minimum 2 words
    workers=4,
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs) # default epochs is 5 in model.epochs
```
