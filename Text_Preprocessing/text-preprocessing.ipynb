# **3. Remove URLs**

````python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
````

# **4. Remove Punctuation** <br>
Here is the full list of punctuation marks detected by Python's `string.punctuation`:

```

```

These include common punctuation symbols like periods (`.`), commas (`,`), and question marks (`?`), as well as others such as brackets (`[]`, `{}`), and mathematical symbols (`+`, `*`, `/`).

If I want to not remove '!' this from text so we can do this also.

````python
import string,time
string.punctuation
````

````python
exclude = string.punctuation
````

````python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
````

# **6. Removing Stop Words** 
**example:** a, the, of , are...

````python
from nltk.corpus import stopwords
````

````python
stopwords.words('english')
````

````python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
````

````python
remove_stopwords('probably my all-time favorite movie, a story of selfness, sacrifice and dedication to a noble cause')
````

````python
# faster and optimized code
stop_words = set(stopwords.words('english'))  # Convert to set for faster lookup

def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stop_words])
````

# **7. Handling Emojis**
**Methods:** Remove, Replace <br>
**Example:** 😀 -> Happy (replace)

````python
# remove
import re

def remove_emoji(text):
    emoji_pattern = re.compile("["
                              u"\U0001F600-\U0001F64F"  # Emoticons
                              u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
                              u"\U0001F680-\U0001F6FF"  # Transport & map symbols
                              u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
                              u"\U00002702-\U000027B0"  # Dingbats
                              u"\U000024C2-\U0001F251"  # Enclosed characters
                              "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
````

# **9. Stemming**
"In grammar, inflectionis the modification of a word to express different grammatical categories as such as tense. case, voice, aspect, person, number, genderm and mood. <br>
walk -> walking, walked, walks

"Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid words to the same stem even if the stem itself is not a valid word in the language."

### **Using NLTK for Stemming**
2 Types of NLTK 
- Porter stemmer -> English
- Snow board stemmer -> Other Language

````python
from nltk.stem.porter import PorterStemmer
````

````python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
````

````python
sample = "walk walks walking walked"
stem_words(sample)
````

### **Lemmatization**
Lemmatization, unlike stemming, reduce the inflected words properly ensuring that the root word belongs to the language. In Lemmatization root word is called Lemma. A Lemma (Plural lemmas or lemmata) is the canocical form, dictinary form, or citation form of a set of words.

## **Stemming VS Lemmatization**
- Stemming is a Algorithm based -> So,it is a fast process
- Lemmatization is a searching based. Find word in Lexical dictionary. -> So, it is a slow process as compare to stemming.
